```python
#importing libraries
import requests
from bs4 import BeautifulSoup 
import pandas as pd 
from time import sleep
import random
import selenium
from selenium import  webdriver
from selenium.webdriver.common.by import By
```

```python
#open_browser and URL
browser = webdriver.Chrome()
browser.maximize_window()
companies = "https://www.bcorporation.net/en-us/find-a-b-corp/"
browser.get(companies)
sleep(8)
print(browser.title)
```

```python
#Data lists to store extracted info
company_names = [] 
headquarters = []
certified_since = []
industry = []
sector = []
operates_in = [] 
website = []
```

## Scraping Companies from Bcorporation


```python
# Loop through pages and scrape data

while True: 
    print('Scraping page', i + 1)
        
    # Locate all product elements on the page
    company_elements = browser.find_elements(By.CLASS_NAME,"ais-Hits-item")
       # Extract hrefs from <a> tags inside each company card
    company_urls = []
    for elem in company_elements:
        try:
            link = elem.find_element(By.CSS_SELECTOR, 'a[data-testid="profile-link"]')
            href = link.get_attribute("href")
            if href:
                company_urls.append(href)
        except:
            continue

    print(f"Found {len(company_urls)} company links on page {i + 1}")
    
    # Visit each company page by href and extracting data
    for url in company_urls:
        # Open company page in new tab
        browser.execute_script("window.open(arguments[0]);", url)
        browser.switch_to.window(browser.window_handles[1])
        sleep(5)

       


        try:
            name = browser.find_element(By.TAG_NAME, "h1").text
        except:
            name = "N/A"
        try:
            headquarter = browser.find_element(By.XPATH,'//span[text()="Headquarters"]/following-sibling::div/p').text
                                                       
        except:
            headquarter = "N/A"
        try:
            certified = browser.find_element(By.XPATH, '//span[text()="Certified Since"]/following-sibling::div/p').text
        except:
            certified = "N/A"
        try:
            ind = browser.find_element(By.XPATH, '//span[text()="Industry"]/following-sibling::div/p').text
        except:
            ind = "N/A"
        try:
            sec = browser.find_element(By.XPATH, '//span[text()="Sector"]/following-sibling::div/p').text
        except:
            sec = "N/A"
        try:
            operates = browser.find_element(By.XPATH, '//span[text()="Operates In"]/following-sibling::div/p').text
        except:
            operates = "N/A"
        try:
            web = browser.find_element(By.XPATH, '//span[text()="Website"]/following-sibling::div').text
        except:
            web = "N/A"
            

        print("Company Name:",name,"\nHeadquarter:", headquarter, "\nCertified Since:", certified, "\nIndustry:", ind,"\nSector:", sec,"\nweb:",web )
        company_names.append(name)
        headquarters.append(headquarter)
        certified_since.append(certified)
        industry.append(ind)
        sector.append(sec)
        operates_in.append(operates)
        website.append(web)
        browser.close()
        browser.switch_to.window(browser.window_handles[0])
    # Click Next button on listing page
    try:
        next_btn = browser.find_element(By.XPATH, '//*[@id="___next"]/div/main/div[4]/div/nav/button[2]')
        next_btn.click()
        sleep(9)
        i += 1
    except:
        print("No more pages.")
        break
```

```python
#create a dataframe
companies = pd.DataFrame({
    'Company_names' : company_names, 
    'Headquarters' : headquarters,
    'Certified_since' : certified_since,
    'Industry' : industry,
    'Sector' : sector,
    'Operates_in' : operates_in, 
    'Website' : website
    })
companies.to_csv('bcorp_companies.csv',index = False, encoding = 'utf-8-sig')
print('Mission Accomplished file saved successfully')
```
